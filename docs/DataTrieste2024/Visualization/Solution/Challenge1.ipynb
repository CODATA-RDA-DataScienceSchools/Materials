1. Remove the legend from this plot



```python
# Loading/installing packages

# Boxplot for life expectancy by continent
p1 = ggplot(gapminder, aes(x='continent', y='lifeExp', fill='continent'))
p1 = p1 + geom_boxplot(outlier_size=2) #Then, add ageom_boxplot() layer
p1 = p1 + guides(fill=False) #Then, remove legend
p1.show()
```

2. Make the plot horizontal

```python
p2 = p1 + coord_flip()
p2.show()
```

3. Instead of a boxplot, try geom_violin()

```python
# Vilolin plot for life expectancy by continent
p3 = ggplot(gapminder, aes(x='continent', y='lifeExp', fill='continent'))
p3 = p3 + geom_violin()
p3.show()
```
