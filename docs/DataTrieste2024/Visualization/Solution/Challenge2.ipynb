1. Plot the distributions of GDP per capita separately for each continent.




```python
# Don't forget to load libraries
# Density plot for gdpcapita by continent
p = (ggplot(gapminder, aes(x='gdpPercap', fill='continent')) +
     geom_density(alpha=0.3))
p.show()
```

2. Transform the x-axis to a log scale and add another layer for the log transformation.

```python
# Density plot for gdpcapita by continent with log scale
p = (ggplot(gapminder, aes(x='gdpPercap', fill='continent')) +
     geom_density(alpha=0.3)) + coord_trans(x='log')
p.show()
```

3. Create boxplots of GDP per capita by continent, with and without a log scale.

```python
# Boxplot for gdpPercap by continent
gap1 = ggplot(gapminder, aes(x='continent', y='gdpPercap', fill='continent'))
p = gap1 + geom_boxplot(outlier_size=2) #Then, add ageom_boxplot() layer:
p.show()
```
