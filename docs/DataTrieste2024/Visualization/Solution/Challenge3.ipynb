Make a plot of life expectancy vs. GDP per capita for China and India, with both lines and points.


```python
# Methode 1
# Filter data for China and India
china_india = gapminder[(gapminder['country'] == 'China') | (gapminder['country'] == 'India')]

china_india_plot = (ggplot(china_india, aes(x='gdpPercap', y='lifeExp'))
                          + geom_line()
                          + geom_point())
china_india_plot.show()
```

```python
# Methode 2
# Filter data for China 
china = gapminder[gapminder['country'] == 'China'] 

# Filter data for India
india = gapminder[gapminder['country'] == 'India']

# Concatenate the two DataFrames
china_india = pd.concat([china, india])

china_india_plot = (ggplot(china_india, aes(x='gdpPercap', y='lifeExp'))
                          + geom_line()
                          + geom_point())
china_india_plot.show()
```

```python
# separate lines
china_india_plot = (ggplot(china_india, aes(x='gdpPercap', y='lifeExp', fill='country'))
                          + geom_line()
                          + geom_point())
china_india_plot.show()
```
