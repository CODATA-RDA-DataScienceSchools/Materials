# BRICS World Bank Indicators Analysis

## Data Manipulation

First we read the `csv` file that contains Economical data.

```r
economy_data <- read.table("Economy_Data.csv", sep=";", header=T)
```

```r
head(economy_data)
```

Let's take a closer look at the economical indicators that we have inside our dataset

```r
unique(economy_data[1])
```

```r
library(data.table)
```

```r
unique(economy_data[economy_data$SeriesName %like% "GDP",][1])
```

Select the data about the GDP percapita

```r
gdp_percapita_data <- economy_data[economy_data$SeriesName == 'GDP per capita (constant 2010 US$)',]
```

```r
head(gdp_percapita_data)
```

Drop columns that won't help in our analysis

```r
gdp_percapita_data <- gdp_percapita_data[,c("CountryCode", "Year", "Value")]
```

```r
head(gdp_percapita_data)
```

Let's run a `summary` of our dataset

```r
summary(gdp_percapita_data)
```

Let's remove the **NAs** from our dataset

```r
gdp_percapita_data <- na.omit(gdp_percapita_data)
```

```r
summary(gdp_percapita_data)
```

Now it is time to show how the gdp percapita evolver from 1970 to 2019 in the BRICS countries

```r
library(ggplot2)
```

```r
ggplot(data=gdp_percapita_data, aes(x=Year, y=Value, group=CountryCode)) +
  geom_line(aes(color=CountryCode))
```
