```python
continents = ('Africa','Americas','Asia','Europe', 'Oceania')
medianLifeExps = {}
for val in continents:
    key = gapminder[gapminder.continent==val].lifeExp.median()
    medianLifeExps[key] = val
print(medianLifeExps)    
sortedKeys = sorted(medianLifeExps)  
print(sortedKeys)
sortedContinents = [medianLifeExps[x] for x in sortedKeys]
sortedContinents
```

```python
sns.boxplot(x="continent", y="lifeExp", data=gapminder.sort_values("continent"))['=]']
```

```python
sortedGapminder = pd.DataFrame()
for c in sortedContinents:
    tmpDF = gapminder[gapminder.continent==c]
    sortedGapminder = pd.concat([sortedGapminder, tmpDF], axis=0)
    
    
```
